```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
file = '/content/drive/MyDrive/projects/heart_attack_youngsters_india.csv'
data = pd.read_csv(file)
```

```python
# Display the first few rows of the dataset
print("Dataset Preview:")
data.head()
```

```python
# Display dataset information
print("\nDataset Information:")
data.info()
```

```python
# Check for missing values
print("\nMissing Values:")
data.isnull().sum()
```

```python
# Statistical summary of the dataset
print("\nStatistical Summary:")
data.describe()
```

```python
# Visualize the distribution of numerical columns
numerical_columns = data.select_dtypes(include=[np.number]).columns.tolist()
for col in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Correlation heatmap with numerical columns only
numerical_data = data.select_dtypes(include=[np.number])

if not numerical_data.empty:
    plt.figure(figsize=(10, 6))
    sns.heatmap(numerical_data.corr(), annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Correlation Heatmap")
    plt.show()
else:
    print("No numerical columns available for correlation heatmap.")
```

```python
# Countplot for categorical features
categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
for col in categorical_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(data[col], palette='viridis')
    plt.title(f'Countplot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```
